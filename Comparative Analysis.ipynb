```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
plt.figure(figsize=(8,8))
# Make a fake dataset:
height = [96.7, 94.3, 96.1, 94.9]
bars = ('DT(96.7)', 'NB(94.3)', 'RF(96.1)', 'SVM(94.9)')
y_pos = np.arange(len(bars))

barWidth = 0.6

# Create bars
plt.bar(y_pos, height, color=['#DC353A', '#9B4CEA', '#3CB851', '#DF5F2F'], width=barWidth, edgecolor='white')
plt.title('Accuracy Comparison', color = '#F55200', fontsize='25')
plt.ylabel('Accuracy', color = 'red', fontsize='17')
plt.xlabel('Classification Algorithms', color = 'red', fontsize='17', horizontalalignment='center')

# Create names on the x-axis
plt.xticks(y_pos, bars)
 
# Show graphic
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(9,9))
import seaborn as sns
sns.set()

# set width of bar
barWidth = 0.25

# set height of bar
precision = [0.89, 0.84, 0.88, 0.80]
recall = [0.95, 0.73, 0.95, 1.00]
f1_score = [0.92, 0.78, 0.91, 0.89]

# Set position of bar on X axis
r1 = np.arange(len(precision))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# Make the plot
plt.bar(r1, precision, color='#DC353A', width=barWidth, edgecolor='white', label='precision')
plt.bar(r2, recall, color='#9B4CEA', width=barWidth, edgecolor='white', label='recall')
plt.bar(r3, f1_score, color='#3CB851', width=barWidth, edgecolor='white', label='f1_score')
plt.title('Precision Recall F1-Score Comparison', color = '#F55200', fontsize='20')

plt.ylabel('Score', color = '#B90000', fontsize='17')
plt.xlabel('Classification Algorithms', color = '#B90000', fontsize='17', horizontalalignment='center')
# Add xticks on the middle of the group bars
plt.xticks([r + barWidth for r in range(len(bars1))], ['DT (prec, rec, f1)', 'NB (prec, rec, f1)', 'RF (prec, rec, f1)', 'SVM (prec, rec, f1)'])

# Create legend & Show graphic
plt.legend()
plt.show()
#fig=plt.figure(figsize=(50, 20), dpi= 80, facecolor='w', edgecolor='k')
```
